#### Problem 3

a) Visualize each bivariate relationship among Average_Class_Size, Average_Salary, Perc_Economically_Disadvantaged, and Perc_English_Language_Learner.  Use the Seaborn function **pairplot**.  The argument to `pairplot` should be `school_data[['Average_Class_Size', 'Average_Salary', 'Perc_Economically_Disadvantaged', 'Perc_English_Language_Learner']]`, which will just pull those columns out from the full data frame.

b) Describe what you see from the plots.

c) Based on your observations from 3b, and as a completely post-hoc, exploratory analysis, choose one of the 4 measures from 3a to be an outcome variable, and a second measure to be a predictor variable.  Then run a linear regression, print the *summary*, and write a sentence interpreting the results (does not need to be detailed, just practice retrieving the relevant info on predictor significance, etc.)

d) Commit your changes and push to GitHub!

```python
# Problem 3 code here (can split into multiple code chunks if you want)
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# a)
sns.pairplot(school_data[['Average_Class_Size', 'Average_Salary', 'Perc_Economically_Disadvantaged', 'Perc_English_Language_Learner']])
# b)
#Percent ELL seems positively correlated with Percent economically disadvantaged, average salary, and average class size, but average class size and salary don't seem to be correlated with each other.
# c)
school_reg = smf.ols('Average_Class_Size ~ Perc_English_Language_Learner', data=school_data).fit()
print(school_reg.summary())
#For every increase in 1 percent english language learners, the predicted average class size increases by 0.03 people on average; the p value is less than 0.01, indicating this is a significant relationship.
```

#### Problem 4

a) Create a new linear regression model that takes the model from 3c and adds both of the remaining variables from 3a as additional predictor variables.  To do this, the formula interface would look like `"outcome_name ~ predictor1_name + predictor2_name + predictor3_name"`

b) Print the summary from 4a, and write a short interpretation of the results, especially comparing it to Problem 3c.  

c) Add School_Type as another predictor to your model.  Interpret the new predictor in your output.  How is it different from the other predictors?

d) Commit your changes and push to GitHub!

```python
# Problem 4 code here (can split into multiple code chunks if you want)
# a)
school_reg4 = smf.ols('Average_Class_Size ~ Perc_English_Language_Learner + Perc_Economically_Disadvantaged + Average_Salary', data=school_data).fit()
# b)
print(school_reg4.summary())
#In this model, Percent ELL and Percent economically disadvantaged are both significant predictors for average class size, but average salary is not. 
# The coefficient also increased from 0.03 to 0.08, indicating that an increase in percent ELL is associated with a *greater* predicted increase in average class size when accounting for percent economically disadvantaged and average salary.
# c)
school_regc = smf.ols('Average_Class_Size ~ Perc_English_Language_Learner + Perc_Economically_Disadvantaged + Average_Salary + School_Type', data=school_data).fit()
print(school_regc.summary())
# School Type is a categorical variable, as compared to the other variables which are numeric/continuous.
# As compared to private schools, the predicted average class size for public schools will have 9.8 more people (holding all other variables constant).
```
